```python
## Importing the CSV and creation of a dataframe "poke_df"
import pandas as pd
from matplotlib import pyplot as plt

PATH_POKEDEX_CSV = "./pokedex/pokedex.csv"
poke_df = pd.read_csv(PATH_POKEDEX_CSV)
```

```python
## Listing all the attributes/colname in the dataframe, with their respective datatypes
print(poke_df.columns)
```

```python
## "Unnamed: 0" is not useful for our case, so we remove the column
poke_df = poke_df.drop("Unnamed: 0", axis=1)
poke_df.head()
```

```python
## Size of the DataFrame: There are 1028 different Pokemons
shape_poke_df = poke_df.shape
print(f"Rows: {shape_poke_df[0]}\nColumns: {shape_poke_df[1]}")
```

```python
## Generations of Pokemon: 8 (2024)
## According to Bulbapedia: A generation (Japanese: 世代 ( せだい ) generation) is a grouping of the Pokémon games that separates them based on the Pokémon they include.

generations = poke_df["generation"].value_counts().sort_index()
print(generations)
```
